```python
df_train = pd.read_csv("train.csv")
```

```python
df_train_r = df_train
```

```python
del df_train_r["PassengerId"]
del df_train_r["Embarked"]
del df_train_r["Name"]
del df_train_r["Ticket"]
df_train_r['Cabin'] = df_train_r['Cabin'].fillna(0)
df_train_r['Cabin'] = df_train_r['Cabin'].astype(str).str[0]
df_train_r['Age'] = df_train_r['Age'].fillna(0)
```

```python
df_train_r
```

```python
def age_group(x):
    if x ==0:
        return('Age_Null')
    elif x > 0 and x <= 10:
        return ('Age1to10')
    elif x >10 and x <= 20:
        return ('Age1to20')
    elif x >20 and x <= 30:
        return ('Age1to30')
    elif x >30 and x <= 40:
        return ('Age1to40')
    elif x >40 and x <= 50:
        return ('Age1to50')
    elif x >50 and x <= 60:
        return ('Age1to60')
    elif x >60 and x <= 70:
        return ('Age1to70')
    elif x >70 and x <= 80:
        return ('Age1to80')
    elif x >80 and x <= 90:
        return ('Age1to90')
    elif x >90 and x <= 100:
        return ('Age1to100')
```

```python
df_train_r['Age'] = df_train_r['Age'].apply(lambda x:age_group(x))
```
