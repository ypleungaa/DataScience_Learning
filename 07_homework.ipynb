```python
# By Wayne
#Part 1

hw1 = df_flights.groupby(['AIRLINE','ORIGIN_AIRPORT']).agg({
    'FLIGHT_NUMBER': 'count', ## only under the panda library , so we can just input a string then help us to coun == lambda x: x.count()
    'OPERATED RATE' : 'mean', #similar as above / below
    'DELAY RATE': lambda x: x.mean(),
})
hw1
```

```python
# We only need those delayed
#Part 2
# 1 = delay , 0 = no delay
hw2 = df_flights.loc[df_flights['DELAY RATE'] == 1].groupby(['AIRLINE','ORIGIN_AIRPORT']).agg({
    'DEPARTURE_DELAY': 'mean' # DEPARTURE_DELAY have -ve number !!! ,which is not make sense for delay
})
hw2
```

```python
# Merge two dataframes from Part 1 , Part 2 (similar join in SQL)
#But need to reset the index before merge
```

```python
hw1_r=hw1.reset_index()
hw1_r
```

```python
hw2_r=hw2.reset_index()
hw2_r
```

```python
# similar to SQL, select * from table 1 join table 2 on column 1 = column
hw = hw1_r.merge(hw2_r, on=['AIRLINE', 'ORIGIN_AIRPORT'], how = 'inner') # need to combine all duplicated column for merging, otherwise, it will show both duplicated column in the dataframe
# inner only join rows in both dataframe , refer to the cheat sheet
hw
```
